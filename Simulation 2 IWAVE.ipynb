```python

nit = 50
indiv = 1000
Npop = 1

epochs=10000
batch=64

latent_size = 1
lr = 0.001


random.seed(31)


##############################################################################################################
# Program


a = np.ones((100,nit))
b = np.zeros((100,nit))
thetas = np.zeros((100,indiv))
fitness_list = np.zeros((100,1))



## Replicas
for i in range(100):

    print(f"Replica {i}")
    print("")
        
    tit = "DataEvo/"
    data = pd.read_csv(tit+"Data_"+str(i+1)+".csv")
    
    Y = torch.from_numpy(data.to_numpy())
##############################################################################################################
    # Network  
    model = IWAVE(model_type = "grm",
              learning_rate = lr,
#              device = device,
              inference_net_sizes = [20],
              latent_size = latent_size,
              n_cats = [2] * nit,
             )

    model.fit(Y, batch_size = 64, iw_samples = 10)

    
    
    a[i] = np.around(model.loadings.numpy(), 2).flatten()
    b[i] = np.around(model.intercepts.numpy(), 2).flatten()
    thetas[i] = model.scores(Y).flatten()
    

# Network outputs 
a = pd.DataFrame(np.transpose(a))
b = pd.DataFrame(np.transpose(b))
thetas = pd.DataFrame(np.transpose(thetas))


```

```python
a.to_csv('OutputsEvo/aIWAVE.csv')
b.to_csv('OutputsEvo/bIWAVE.csv')
thetas.to_csv('OutputsEvo/thetasIWAVE.csv')
```
